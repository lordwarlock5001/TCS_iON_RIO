```python
dataset["is_blue"]=''
for i in range(len(dataset)):
    if dataset['blue'][i]==0:
        dataset.loc[i,'is_blue']='No'
    else:
        dataset.loc[i,'is_blue']='Yes'
dataset
```

```python
dg=dataset.groupby('is_blue')
dg=dg.count()
plt.bar(dg.index,dg.blue)
```

there are approximately the same numnber 

```python
dataset["cores"]=''
for i in range(len(dataset)):
    if dataset['n_cores'][i]==1:
        dataset.loc[i,'cores']='single core'
    elif dataset['n_cores'][i]==2:
        dataset.loc[i,'cores']='dual core'
    elif dataset['n_cores'][i]==3:
        dataset.loc[i,'cores']='tri core'
    elif dataset['n_cores'][i]==4:
        dataset.loc[i,'cores']='quad core'
    elif dataset['n_cores'][i]==5:
        dataset.loc[i,'cores']='5 core'
    elif dataset['n_cores'][i]==6:
        dataset.loc[i,'cores']='hexa core'
    elif dataset['n_cores'][i]==7:
        dataset.loc[i,'cores']='7 core'
    elif dataset['n_cores'][i]==8:
        dataset.loc[i,'cores']='octa core'
dg=dataset.groupby('cores')
dg=dg.count()
fig = plt.figure(figsize = (10, 5))
plt.bar(dg.index,dg.blue)
```
